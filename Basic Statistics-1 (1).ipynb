Standardization of Numerical Variables

Standardization (z-score normalization) is a process that involves rescaling data so that it has a mean of 0 and a standard deviation of 1. This transformation is essential for comparing data across different scales and for preparing data for statistical analysis and machine learning. It involves calculating the z-score for each data point, which tells you how many standard deviations away from the mean the point is.

```python
from sklearn.preprocessing import StandardScaler , MinMaxScaler
```

```python
Sc = StandardScaler()
```
